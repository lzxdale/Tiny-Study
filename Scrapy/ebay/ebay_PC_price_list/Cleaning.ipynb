```python
leno = {'T':[],'E':[],'T_sum':0,'E_sum':0}
for i in file_Len.values:
    if re.search(r"T\d\d\d",i[2].upper()):
        price = float(''.join(i[3][1:].split(',')))
        sec_info = i[4]
        year = i[0]
        brand = i[1]
        if price < 3000:
            leno['T'].append([brand,'T-series',i[2],price,sec_info,year])
            leno['T_sum'] += price
    elif re.search(r"E\d\d\d",i[2].upper()):
        price = float(''.join(i[3][1:].split(',')))
        sec_info = i[4]
        year = i[0]
        brand = i[1]
        if price < 3000:
            leno['E'].append([brand,'E-series',i[2],price,sec_info,year])
            leno['E_sum'] += price
```

```python
Dell = {'7':[],'5':[],'7_sum':0,'5_sum':0}
for i in file_Dell.values:
    if re.search("latitude", i[2].lower()): #first check if its a latitude
        if re.search(r"7\d\d\d",i[2].lower()):
            price = float(''.join(i[3][1:].split(',')))
            sec_info = i[4]
            year = i[0]
            brand = i[1]
            if price < 3000:
                Dell['7'].append([brand,'7-series',i[2],price,sec_info,year])
                Dell['7_sum'] += price
        elif re.search(r"5\d\d\d",i[2].lower()):
            price = float(''.join(i[3][1:].split(',')))
            sec_info = i[4]
            year = i[0]
            brand = i[1]
            if price < 3000:
                Dell['5'].append([brand,'5-series',i[2],price,sec_info,year])
                Dell['5_sum'] += price
```

```python
HP = {'Probook':[],'Elitebook':[],'P_sum':0,'E_sum':0}
for i in file_HP.values:
    if re.search(r"probook",i[2].lower()):
        price = float(''.join(i[3][1:].split(',')))
        sec_info = i[4]
        year = i[0]
        brand = i[1]
        if price < 3000:
            HP['Probook'].append([brand,'Probook',i[2],price,sec_info,year])
            HP['P_sum'] += price
    elif re.search(r"elitebook",i[2].lower()):
        price = float(''.join(i[3][1:].split(',')))
        sec_info = i[4]
        year = i[0]
        brand = i[1]
        if price < 3000:
            HP['Elitebook'].append([brand,'Elitebook',i[2],price,sec_info,year])
            HP['E_sum'] += price
```

```python
leno_E_series_avg_pr = leno['E_sum']/len(leno['E'])
leno_T_series_avg_pr = leno['T_sum']/len(leno['T'])
Dell_7_series_avg_pr = Dell['7_sum']/len(Dell['7'])
Dell_5_series_avg_pr = Dell['5_sum']/len(Dell['5'])
HP_7_series_avg_pr = HP['P_sum']/len(HP['Probook'])
HP_5_series_avg_pr = HP['E_sum']/len(HP['Elitebook'])
```

```python
import csv
avg_price = [leno_E_series_avg_pr,
            leno_T_series_avg_pr,
            Dell_7_series_avg_pr,
            Dell_5_series_avg_pr,
            HP_7_series_avg_pr,
            HP_5_series_avg_pr]
name = ['leno_E_series_avg_pr',
        'leno_T_series_avg_pr',
        'Dell_7_series_avg_pr',
        'Dell_5_series_avg_pr',
        'HP_7_series_avg_pr',
        'HP_5_series_avg_pr']

def write_info(writer, the_list):
    for i in list(the_list.values())[0]:
        writer.writerow(['',i[0],i[1],i[2],'$'+str(i[3]),i[4],i[5]])
    for i in list(the_list.values())[1]:
        writer.writerow(['',i[0],i[1],i[2],'$'+str(i[3]),i[4],i[5]])
    
        
with open('PC_ebay_list.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['Avg_price_info'])
    for p, n in zip(avg_price, name):
        writer.writerow(['',n,'$'+str(p)])
    writer.writerow(['DataList','Brand', 'Series', 'name', 'price', 'second_info','year'])
    for thelist in [leno,Dell,HP]:
        write_info(writer,thelist)
```
