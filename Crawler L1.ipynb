```python
def requests_view(response):
    import webbrowser
    request_url = response.url
    base_url = '<head><base href="%s">'%(request_url)
    base_url = base_url.encode()
    content = response.content.replace(b"<head>",base_url)
    tem_html = open('tmp.html','wb')
    tem_html.write(content)
    tem_html.close()
    webbrowser.open_new_tab("tmp.html")
```

```python
response = requests.get("http://www.spbeen.com/tool/request_info/")
requests_view(response)
```

```python
base_url = "http://jandan.net/pic/page-{}"
num = 1
while True:
    time.sleep(0.2)
    print('current pages: {}'.format(num))
    status = pass_page(base_url.format(num),num)
    if status:
        num += 1
    else:
        break
print('End')
```
