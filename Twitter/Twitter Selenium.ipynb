## Using Selenium to access and download tweets.

```python
from selenium import webdriver
opt = webdriver.ChromeOptions()
opt.binary_location = "/Applications/Chrome.app/Contents/MacOS/Google Chrome"
```

### Aviod loading picture, fasten the speed

```python
prefs = {}
prefs['profile.managed_default_content_settings.images'] = 2
opt.add_experimental_option('prefs',prefs)
```

```python
drive = webdriver.Chrome(options = opt)
vix_url = 'https://twitter.com/search?q=%23vix&src=typd'
drive.get(vix_url)
```

## First, just downloading text, no images

```python
from datetime import datetime
import pandas as pd
import time
```

```python
for i in range(300):
    drive.execute_script("window.scrollTo(0, document.body.scrollHeight)") # scrolling down pages
    time.sleep(1.5)
```

```python
all_time = drive.find_elements_by_xpath('.//div[@class="content"]//small[@class="time"]/a/span')
all_text = drive.find_elements_by_xpath('.//div[@class="js-tweet-text-container"]//p')
all_lang = drive.find_elements_by_xpath('.//div[@class="js-tweet-text-container"]//p')
all_name = drive.find_elements_by_xpath('.//div[@class = "stream-item-header"]/a/span[1]')
all_id = drive.find_elements_by_xpath('.//div[@class = "stream-item-header"]/a/span[2]')
time_list = []
for i in all_time:
    eachtime = i.get_attribute('data-time')
    if eachtime:
        time_list.append(datetime.fromtimestamp(int(eachtime)))
name_list = []
id_list = []
text_list = []
lang_list = [] ##record all langages, and later just keep english. 
for i in range(len(all_name)):
    lang_list.append(all_lang[i].get_attribute('lang'))
    name_list.append(all_name[i].text.split('\n')[0])
    id_list.append(all_id[i].text)
    text_list.append(all_text[i].text)
```

```python
df_msg = pd.DataFrame({'time_list':time_list, 'name_list':name_list, 'id_list':id_list, 'text_list':text_list,'Lang':lang_list})
```
